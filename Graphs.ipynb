```python
import pandas as pd
#first dataset
d2={'count': [57.25, 63.36], 'count stopwords': [68.97, 63.395],
   'tfidf': [37.64, 63.30], 'tfidf stopwords': [41.18, 63.404]}

df2=pd.DataFrame(d2)
df2
```

```python
df2['Naive Bayes']=['Multinomial', 'Gaussian']
df2=df2.set_index('Naive Bayes')
df2
```

```python
import matplotlib.pyplot as plt
colors={'count': "SkyBlue", 'count stopwords': "#86bf91",'tfidf' : 'pink', 'tfidf stopwords': 'plum' }
df2.plot(kind='barh',color=colors, figsize=(6, 6),zorder=2, width=0.5)
plt.title("20newsgroups 5-fold Cross Validation Hyper-Parameters")
plt.xlabel("Validation Accuracy(%)")

plt.xlim(30, 75)
```
